```python
import pandas as pd

annotation = pd.read_excel("D:/Covid_data_New/Yr2023/6Jun/20Jun/AnnotationRoster-BenchMark.xlsx",sheet_name="Sheet1")

```

```python
import os

dataFromFile = dict()

oFile = open("D:/Covid_data_New/Yr2023/6Jun/20Jun/ampSIFDataFormat.txt","w")
for i in annotation.index:
    fileName = annotation.loc[i,'File Name'].strip()
    well = annotation.loc[i,'Well Position']
    target = annotation.loc[i,'Target']
    dye = annotation.loc[i,'Dye']
    prob = annotation.loc[i,'Clean/ Problem']
    dataFromFile.setdefault(fileName,[]).append(well)
    oFile.write(fileName+"\t"+str(well)+"\t"+target+"\t"+dye+"\t"+prob+"\n")

oFile.close()
```

```python
fileList = os.listdir("D:/Covid_data_New/Yr2023/6Jun/20Jun/Excels/")
oFile1 = oFile = open("D:/Covid_data_New/Yr2023/6Jun/20Jun/ampPerfdata.txt","w")
oFile1.write("File Name\tWell Id\tTarget\n")

for excelFiles in fileList:
    splitName = excelFiles.split("_20230620")
    runCode = True
    try:
        retVal = dataFromFile[splitName[0]]
        
    except (KeyError):
        runCode=False

    if runCode == False:
        exit()
    else:

        daExport = pd.read_excel("D:/Covid_data_New/Yr2023/6Jun/20Jun/Excels/"+excelFiles,sheet_name="Amplification Data"
                            ,skiprows=20)
        if dataFromFile[splitName[0]]:
            vals = dict()
            for i in daExport.index:
                well = daExport.loc[i,"Well"]
                for wells in dataFromFile.get(splitName[0]):
                    if well == wells:
                        target = daExport.loc[i,"Target"]
                        dRN = daExport.loc[i,"dRn"]
                        comb = splitName[0]+"\t"+str(well)+"\t"+target
                        vals.setdefault(comb,[]).append(dRN)
        
            for key in vals:
                oFile1.write(key+"\t")
                for dRnValue in vals.get(key):
                    oFile1.write(str(dRnValue)+"\t")
                oFile1.write("\n")
                #oFile1.flush() 
    
oFile1.close()
```
