```python
import numpy as np
import pandas as pd

def getInformation(metaData,completeData1,listName):
    dataForInfo = []
    dataFrame = pd.DataFrame()
    j=0
    for i in metaData[0]:
        if i == "File Name":
            fileN = metaData[1][j].split("\\")
            ff = metaData[1][j]
            #print(ff[0:ff.rindex("\\")],fileN[-1])
            #fileWrite.write(ff[0:ff.rindex("\\")]+"\t"+fileN[-1]+"\n")
            dataForInfo.append(ff[0:ff.rindex("\\")])
            dataForInfo.append(fileN[-1])
        if i == "Instrument Type":
            dataForInfo.append(metaData[1][j])
        if i == "Block Type":
            dataForInfo.append(metaData[1][j])
        if i == "Passive Reference":
            dataForInfo.append(metaData[1][j])      
        j+=1
    
    
    sampleList = []
    targetDayes = []
    for i in completeData1.index:
        samples = completeData1.loc[i,"Sample"]
        targetDayes.append(completeData1.loc[i,"Reporter"])
        sampleList.append(samples)

    dyes = np.unique(targetDayes)
    totalSamples = len(sampleList)/len(dyes)
    dataForInfo.append(totalSamples)
    listDyes = ""
    for j in range(0,len(dyes)-1):
        listDyes += dyes[j]+"/"
    listDyes += dyes[-1]
    dataForInfo.append(listDyes)
    dataFrame = pd.DataFrame([dataForInfo],[listName],["Folder Path", "File Name","Instrument Type","Block Type","Passive Reference","Total Samples","Reporter"])
    return dataFrame    
```

```python
from os import listdir
outFolder = "D:/Covid_data_New/Yr2023/Feb/Feb27/"
bigDF = pd.DataFrame()
inputFolder = "D:/Covid_data_New/Yr2023/Feb/Feb27/Excels/"
listName=0
filesInDir = listdir(inputFolder)
listExcel = []
for file in filesInDir:
    if "xlsx" in file:
            listName+=1
            annotationFile = pd.read_excel(inputFolder+"/"+file,sheet_name="Results",header=None)
            splice = 0
            for j in annotationFile.index:
                if annotationFile.iloc[j][1] == "Well Position":
                    splice = j
                    break
            #print(splice)
            #print(annotationFile)
                  
            #print(file+"\t"+s)
            #outFile.write(file+"\t"+str(annotationFile.loc[1,"Sample"])+"\n")
            metaData = annotationFile[:splice]
            completeData = annotationFile[splice:]
            completeData1 = pd.DataFrame(completeData.values[1:],columns= completeData.iloc[0])
            bigDF = pd.concat([getInformation(metaData,completeData1,listName),bigDF])
bigDF = bigDF[::-1]
excelWriter = pd.ExcelWriter(outFolder+"/Viper_Export_Summary_Matrix_02282023.xlsx")
bigDF.to_excel(excel_writer=excelWriter,sheet_name= "Data",index=False)
excelWriter.close()
```
