Plotting in python

```python
import matplotlib.pyplot as plt
import pandas as pd
mypath = "D:/Covid_data_New/Yr2023/Jan/Jan31/20230130_SFSH_LOD_day1_NP Plate1_QS5_0044_20230130_142341.xlsx"
exportFile = pd.read_excel(mypath,sheet_name="Amplification_Data",skiprows=26)
sampleTarget = dict()
for i in exportFile.index:
    sample = exportFile.loc[i,"Sample"]
    target = exportFile.loc[i,"Target"]
    string = sample+"\t"+target
    cycleNumber = exportFile.loc[i,"Cycle Number"]
    dRn = exportFile.loc[i,"dRn"]
    cycleValue = str(cycleNumber)+"\t"+str(dRn)
    sampleTarget.setdefault(string,[]).append(cycleValue)

#plt.style.use('ggplot')
print(plt.style.available)
for i in sampleTarget:
    x=[]
    y=[]
    for v in sampleTarget[i]:
        z = v.split("\t")
        x.append(z[0])
        y.append(z[1])
    plt.plot(x,y)
    #plt.ylim(-10000, 40000)
    plt.show()
        

#price = [2.50, 1.23, 4.02, 3.25, 5.00, 4.40]
#sales_per_day = [34, 62, 49, 22, 13, 19]

#plt.scatter(price, sales_per_day)
#plt.show()
```
